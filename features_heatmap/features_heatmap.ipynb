**Feature heatmaps with Pivot tables**

https://allisonhorst.github.io/palmerpenguins/

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/penguins_size.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df_p = df.pivot_table(index='species', columns='island', values='flipper_length_mm')
```

```python
sns.heatmap(df_p, cmap='coolwarm')
```
