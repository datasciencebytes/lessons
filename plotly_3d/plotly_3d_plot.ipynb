**Interactive 3D Plot with Plotly**

```python
# !pip install Plotly
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get some data to plot
data = np.array([np.random.randint(1, 300, 100),
                 np.random.randint(1, 500, 100),
                 np.random.randint(1, 100, 100)]
            ).T
classes = np.array(np.random.randint(0, 2, 100))
```

```python
data.shape, classes.shape
```

```python
# plot a 3d plot using plt
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d')

scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=classes)
ax.set_title('My first 3D plot')
```

```python
# plot an Interactive 3d plot using Plotly
```

```python
import plotly.graph_objs as go
import plotly.io as pio
```

```python
data.shape, classes.shape
```

```python
# plot with Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=data[:, 0],
    y=data[:, 1],
    z=data[:, 2],
    mode='markers',
    marker=dict(
        size=3,
        color=classes
    )
)

]
                )

fig.show()
```

```python
# save file to disk
file_name = 'figure.html'
pio.write_html(fig, file_name)
```
