**3D Plot with Matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get some data to plot
data = np.array([np.random.randint(1, 300, 100),
                 np.random.randint(1, 500, 100),
                 np.random.randint(1, 200, 100),
]   
).T
classes = np.array(np.random.randint(0, 2, 100))
```

```python
data.shape, classes.shape
```

```python
# plot a 3d plot using plt
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d')

scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=classes)
ax.set_title('My first 3D plot')
```
